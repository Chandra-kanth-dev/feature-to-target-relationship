```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns


df['pickup'] = pd.to_datetime(df['pickup'])
df['dropoff'] = pd.to_datetime(df['dropoff'])

# ===============================
# 2. FEATURE ENGINEERING (ONLY PICKUP & DROPOFF)
# ===============================
df['pickup_hour'] = df['pickup'].dt.hour
df['dropoff_hour'] = df['dropoff'].dt.hour
df['pickup_quarter'] = df['pickup'].dt.quarter

df['pickup_dayname'] = df['pickup'].dt.day_name()
df['dropoff_dayname'] = df['dropoff'].dt.day_name()

df['pickup_is_weekend'] = df['pickup_dayname'].isin(['Saturday', 'Sunday'])

# Trip duration (minutes)
df['trip_duration'] = (df['dropoff'] - df['pickup']).dt.total_seconds() / 60

# Remove invalid trips
df = df[df['trip_duration'] > 0]

# ===============================
# 3. BASIC ANALYSIS
# ===============================
print("\nPickup Hour Summary:")
print(df['pickup_hour'].describe())

print("\nDropoff Hour Summary:")
print(df['dropoff_hour'].describe())

print("\nTrip Duration Summary (minutes):")
print(df['trip_duration'].describe())

# ===============================
# 4. VISUALIZATION
# ===============================
sns.histplot(df['pickup_hour'], bins=24)
plt.title("Pickup Hour Distribution")
plt.show()

sns.histplot(df['dropoff_hour'], bins=24)
plt.title("Dropoff Hour Distribution")
plt.show()

sns.scatterplot(x='pickup_hour', y='trip_duration', data=df)
plt.title("Pickup Hour vs Trip Duration")
plt.show()

# ===============================
# 5. HYPOTHESIS TESTS
# ===============================

# ------------------------------------------------
# TEST 1: Pickup Hour vs Dropoff Hour (Paired t-test)
# H0: Mean pickup hour = mean dropoff hour
# ------------------------------------------------
t_stat, p_val = stats.ttest_rel(df['pickup_hour'], df['dropoff_hour'])

print("\nHYPOTHESIS TEST 1: Pickup Hour vs Dropoff Hour")
print("T-statistic:", t_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("âœ… Reject H0 â†’ Pickup and dropoff hours differ significantly")
else:
    print("âŒ Fail to reject H0")

# ------------------------------------------------
# TEST 2: Weekend vs Weekday Pickup Duration
# H0: Mean trip duration same
# ------------------------------------------------
weekend = df[df['pickup_is_weekend']]['trip_duration']
weekday = df[~df['pickup_is_weekend']]['trip_duration']

t_stat, p_val = stats.ttest_ind(weekend, weekday, equal_var=False)

print("\nHYPOTHESIS TEST 2: Weekend vs Weekday Trip Duration")
print("T-statistic:", t_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("âœ… Reject H0 â†’ Pickup day affects trip duration")
else:
    print("âŒ Fail to reject H0")

# ------------------------------------------------
# TEST 3: Pickup Hour vs Dropoff Hour Correlation
# H0: No correlation
# ------------------------------------------------
corr, p_val = stats.pearsonr(df['pickup_hour'], df['dropoff_hour'])

print("\nHYPOTHESIS TEST 3: Pickup Hour vs Dropoff Hour Correlation")
print("Correlation:", corr)
print("P-value:", p_val)

if p_val < 0.05:
    print("âœ… Significant correlation exists")
else:
    print("âŒ No significant correlation")

# ------------------------------------------------
# TEST 4: Pickup Hour Effect on Trip Duration (ANOVA)
# H0: Trip duration same for all pickup hours
# ------------------------------------------------
groups = [df[df['pickup_hour'] == h]['trip_duration'] for h in range(24)]

f_stat, p_val = stats.f_oneway(*groups)

print("\nHYPOTHESIS TEST 4: Pickup Hour vs Trip Duration (ANOVA)")
print("F-statistic:", f_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("âœ… Reject H0 â†’ Pickup hour affects trip duration")
else:
    print("âŒ Fail to reject H0")

print("\nâœ… PICKUP & DROPOFF ANALYSIS COMPLETED")
```

```python
import pandas as pd
from scipy import stats

# ===============================
# 1. NUMERIC COLUMNS â†’ CORRELATION TEST
# ===============================
numeric_cols = [
    'pickup_day',
    'pickup_year',
    'dropoff_day',
    'dropoff_year',
    'dropoff_hour',
    'dropoff_quarter',
    'dropoff_second'
]

print("\nðŸ“Š NUMERIC FEATURE SIGNIFICANCE (Pearson Test)\n")

significant_numeric = []

for col in numeric_cols:
    corr, p_val = stats.pearsonr(df[col], df['tip'])
    print(f"{col:20} | Corr = {corr:.3f} | P-value = {p_val:.5f}")

    if p_val < 0.05:
        significant_numeric.append(col)

# ===============================
# 2. CATEGORICAL COLUMNS â†’ ANOVA
# ===============================
categorical_cols = [
    'pickup_month',
    'dropoff_month',
    'dropoff_dayname'
]

print("\nðŸ“Š CATEGORICAL FEATURE SIGNIFICANCE (ANOVA)\n")

significant_categorical = []

for col in categorical_cols:
    groups = [df[df[col] == g]['tip'] for g in df[col].unique()]
    f_stat, p_val = stats.f_oneway(*groups)

    print(f"{col:20} | F-stat = {f_stat:.2f} | P-value = {p_val:.5f}")

    if p_val < 0.05:
        significant_categorical.append(col)

# ===============================
# 3. FINAL RESULT
# ===============================
print("\nâœ… SIGNIFICANT PICKUP / DROPOFF FEATURES FOR TIP PREDICTION\n")

print("ðŸ”¹ Numeric:")
print(significant_numeric if significant_numeric else "None")

print("\nðŸ”¹ Categorical:")
print(significant_categorical if significant_categorical else "None")
```

```python
from scipy import stats

alpha = 0.05
target = 'tip'

columns_to_delete = []

print("\n===================================================")
print("HYPOTHESIS TESTING TO JUSTIFY COLUMN DELETION")
print("Target Variable: tip")
print("Significance Level (alpha):", alpha)
print("===================================================\n")

# --------------------------------------------------
# INDEPENDENT T-TEST FOR YEAR COLUMNS
# --------------------------------------------------
def independent_t_test(col):
    unique_vals = sorted(df[col].unique())

    if len(unique_vals) < 2:
        print(f"Column: {col}")
        print("Not enough groups for testing â†’ MARK FOR DELETION\n")
        columns_to_delete.append(col)
        return

    group1 = df[df[col] == unique_vals[0]][target]
    group2 = df[df[col] == unique_vals[-1]][target]

    t_stat, p_val = stats.ttest_ind(group1, group2, equal_var=False)

    print(f"Column: {col}")
    print("Test Used: Independent t-test")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_val:.6f}")

    if p_val >= alpha:
        print("Decision: âœ… Accept Hâ‚€ â†’ No impact on tip â†’ DELETE COLUMN\n")
        columns_to_delete.append(col)
    else:
        print("Decision: âŒ Reject Hâ‚€ â†’ Has impact on tip â†’ KEEP COLUMN\n")


independent_t_test('pickup_year')
independent_t_test('dropoff_year')

# --------------------------------------------------
# ONE-WAY ANOVA FOR MONTH / QUARTER COLUMNS
# --------------------------------------------------
def one_way_anova(col):
    groups = [df[df[col] == g][target] for g in df[col].unique()]

    if len(groups) < 2:
        print(f"Column: {col}")
        print("Not enough groups for testing â†’ MARK FOR DELETION\n")
        columns_to_delete.append(col)
        return

    f_stat, p_val = stats.f_oneway(*groups)

    print(f"Column: {col}")
    print("Test Used: One-Way ANOVA")
    print(f"F-statistic: {f_stat:.4f}")
    print(f"P-value: {p_val:.6f}")

    if p_val >= alpha:
        print("Decision: âœ… Accept Hâ‚€ â†’ No impact on tip â†’ DELETE COLUMN\n")
        columns_to_delete.append(col)
    else:
        print("Decision: âŒ Reject Hâ‚€ â†’ Has impact on tip â†’ KEEP COLUMN\n")


one_way_anova('pickup_month')
one_way_anova('pickup_quarter')
one_way_anova('dropoff_month')
one_way_anova('dropoff_quarter')

# --------------------------------------------------
# FINAL SUMMARY
# --------------------------------------------------
print("\n===================================================")
print("FINAL DELETION DECISION")
print("===================================================\n")

print("âŒ COLUMNS TO BE DELETED (DO NOT PREDICT TIP):")
for col in columns_to_delete:
    print(" -", col)

print("\nâœ… FINAL ACTION:")
print("Remove these columns before building the prediction model.")

print("\n===================================================")
print("HYPOTHESIS TESTING COMPLETED SUCCESSFULLY")
print("===================================================")
```

```python
from scipy import stats

alpha = 0.05
target = 'tip'

delete_cols = []
keep_cols = []

print("\n===================================================")
print("HYPOTHESIS TESTING FOR PICKUP & DROPOFF TIME FEATURES")
print("Target Variable: tip")
print("Significance Level (alpha):", alpha)
print("===================================================\n")

# --------------------------------------------------
# INDEPENDENT T-TEST (numeric â†’ 2 groups)
# --------------------------------------------------
def independent_t_test(col, split_rule):
    group1 = df[split_rule][target]
    group2 = df[~split_rule][target]

    t_stat, p_val = stats.ttest_ind(group1, group2, equal_var=False)

    print(f"Column: {col}")
    print("Test Used: Independent t-test")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_val:.6f}")

    if p_val < alpha:
        print("Decision: âŒ Reject Hâ‚€ â†’ Influences tip â†’ KEEP\n")
        keep_cols.append(col)
    else:
        print("Decision: âœ… Accept Hâ‚€ â†’ Weak alone â†’ KEEP (behavioral feature)\n")
        keep_cols.append(col)


# --------------------------------------------------
# ONE-WAY ANOVA (categorical â†’ multiple groups)
# --------------------------------------------------
def one_way_anova(col):
    groups = [df[df[col] == g][target] for g in df[col].unique()]
    f_stat, p_val = stats.f_oneway(*groups)

    print(f"Column: {col}")
    print("Test Used: One-Way ANOVA")
    print(f"F-statistic: {f_stat:.4f}")
    print(f"P-value: {p_val:.6f}")

    if p_val < alpha:
        print("Decision: âŒ Reject Hâ‚€ â†’ Influences tip â†’ KEEP\n")
        keep_cols.append(col)
    else:
        print("Decision: âœ… Accept Hâ‚€ â†’ Calendar-based â†’ DELETE\n")
        delete_cols.append(col)


print("--------------- PICKUP FEATURES ---------------\n")

# pickup_year (calendar-based)
print("Column: pickup_year")
print("Test Used: Not applicable (single year / calendar feature)")
print("Decision: âœ… Accept Hâ‚€ â†’ No behavioral meaning â†’ DELETE\n")
delete_cols.append('pickup_year')

# pickup_month
one_way_anova('pickup_month')

# pickup_day
independent_t_test('pickup_day', df['pickup_day'] <= 15)

# pickup_dayname
one_way_anova('pickup_dayname')

# pickup_hour
independent_t_test('pickup_hour', df['pickup_hour'] < 12)


print("--------------- DROPOFF FEATURES ---------------\n")

# dropoff_year
print("Column: dropoff_year")
print("Test Used: Not applicable (single year / calendar feature)")
print("Decision: âœ… Accept Hâ‚€ â†’ No behavioral meaning â†’ DELETE\n")
delete_cols.append('dropoff_year')

# dropoff_month
one_way_anova('dropoff_month')

# dropoff_quarter
one_way_anova('dropoff_quarter')

# dropoff_day
independent_t_test('dropoff_day', df['dropoff_day'] <= 15)

# dropoff_dayname
one_way_anova('dropoff_dayname')

# dropoff_hour
independent_t_test('dropoff_hour', df['dropoff_hour'] < 12)

# dropoff_second
independent_t_test(
    'dropoff_second',
    df['dropoff_second'] <= df['dropoff_second'].median()
)

# --------------------------------------------------
# FINAL SUMMARY
# --------------------------------------------------
print("\n===================================================")
print("FINAL FEATURE SELECTION DECISION")
print("===================================================\n")

print("âŒ COLUMNS TO DELETE:")
for col in sorted(set(delete_cols)):
    print(" -", col)

print("\nâœ… COLUMNS TO KEEP FOR TIP PREDICTION:")
for col in sorted(set(keep_cols)):
    print(" -", col)

print("\n===================================================")
print("HYPOTHESIS TESTING COMPLETED SUCCESSFULLY")
print("===================================================")
```
