Hypothesis testing



```python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
from statsmodels.stats.weightstats import ztest
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

H0: distribution of tip for yellow=distribution of tip for green

H1:distribution of tip for yellow!=distribution of tip for green

```python
# Independent T-Test
yellow_tip = df['tip'].loc[df['color']=='yellow']
green_tip = df['tip'].loc[df['color']=='green']
from scipy import stats

t_stat, p_val = stats.ttest_ind(yellow_tip, green_tip)
print("Independent T-Test statistic:", t_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("Reject null hypothesis: Tips for yellow category is different that of green")
else:
    print("Fail to reject null hypothesis: No significant tips difference by color")
```

H0: distribution of tips paid in cash=distribution of tips paid in credit card

H1:distribution of tips paid in cash!=distribution of tips paid in credit card

```python
# Independent T-Test
cash_tip = df['tip'].loc[df['payment']=='cash']
credit_tip = df['tip'].loc[df['payment']=='credit card']
from scipy import stats

t_stat, p_val = stats.ttest_ind(cash_tip, credit_tip)
print("Independent T-Test statistic:", t_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("Reject null hypothesis: Distribution of Tips paid by cash is different that of paid by credit card")
else:
    print("Fail to reject null hypothesis: No significant tips difference by payment mode")
```

```python
sns.countplot(x=df['pickup_borough'],palette='Set1')
```

H0:Mean of all pickup_borough group is same

H1:Mean of anyone of the group is different

```python
manhattan_tip=df['tip'][df['pickup_borough']=="Manhattan"]
queen_tip=df['tip'][df['pickup_borough']=="Queens"]
brooklyn_tip=df['tip'][df['pickup_borough']=="Brooklyn"]
bronx_tip=df['tip'][df['pickup_borough']=="Bronx"]
f_val,p_val=stats.f_oneway(manhattan_tip,queen_tip,brooklyn_tip,bronx_tip)
print("P-value:", p_val)

if p_val < 0.05:
    print("Reject null hypothesis: mean of anyone is different")
else:
    print("Fail to reject null hypothesis: mean of all groups same ")
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import numpy as np

# Combine all tips
tips = np.concatenate([
    manhattan_tip,
    queen_tip,
    brooklyn_tip,
    bronx_tip
])

# Create group labels
groups = (["Manhattan"] * len(manhattan_tip) +
          ["Queens"] * len(queen_tip) +
          ["Brooklyn"] * len(brooklyn_tip) +
          ["Bronx"] * len(bronx_tip))
tukey = pairwise_tukeyhsd(
    endog=tips,
    groups=groups,
    alpha=0.05
)

print(tukey)
```
